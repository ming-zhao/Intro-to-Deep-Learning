```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
%matplotlib inline
```

```python
def func(x):
    return np.exp(-x ** 2) + 3 * np.exp(-(x - 3) ** 2)
x = np.linspace(-5, 10, 100)

rv = norm(2, 2)
plt.plot(x, func(x), label=r"$\tilde{p}(z)$")
plt.plot(x, rv.pdf(x), label=r"$q(z)$")
plt.legend(fontsize=12)
plt.show()
```

```python
n = 100

sample_candidate = rv.rvs(n * 10)
weight = np.squeeze(func(sample_candidate) / rv.pdf(sample_candidate))
weight /= np.sum(weight)
index = np.random.choice(n * 10, n, p=weight)
samples = sample_candidate[index]

plt.plot(x, func(x), label=r"$\tilde{p}(z)$")
plt.hist(samples, density=True, alpha=0.5)
plt.scatter(samples, np.random.normal(scale=.03, size=(n, 1)), s=5, label="samples")
plt.legend(fontsize=12)
plt.show()
```
