```python
import sys
sys.path.append('../Modules')

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# training data
np.random.seed(1234)
x1 = np.random.normal(size=(100, 2))
x1 += np.array([-5, -5])
x2 = np.random.normal(size=(100, 2))
x2 += np.array([5, -5])
x3 = np.random.normal(size=(100, 2))
x3 += np.array([0, 5])
x_train = np.vstack((x1, x2, x3))

x0, x1 = np.meshgrid(np.linspace(-10, 10, 100), np.linspace(-10, 10, 100))
x = np.array([x0, x1]).reshape(2, -1).T
```

```python
from Mixture import GaussianMixture
gm = GaussianMixture(n_components=3)
gm.fit(x_train)
t = gm.predict_prob(x_train)

plt.scatter(x_train[:, 0], x_train[:, 1], c=t)
plt.scatter(gm.mu[:, 0], gm.mu[:, 1], s=200, marker='X', lw=2, c=['red', 'green', 'blue'], edgecolor="white")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.gca().set_aspect("equal")
plt.show()
```

```python
from sklearn import mixture
gm = mixture.GaussianMixture(n_components=3, covariance_type='full')
gm.fit(x_train)
t = gm.predict(x_train)

plt.scatter(x_train[:, 0], x_train[:, 1], c=t)
plt.scatter(gm.means_[:,0], gm.means_[:, 1], s=200, marker='X', lw=2, c=['red', 'green', 'blue'], edgecolor="white")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.gca().set_aspect("equal")
plt.show()
```
